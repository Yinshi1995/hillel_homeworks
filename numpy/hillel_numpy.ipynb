```python
# Importing libs, fetching data

import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
matrix = np.genfromtxt(url, delimiter=',', dtype='float16')

{'matrix': matrix}
```

```python

# Defining and removing target

target = np.isnan(matrix)
mask = np.invert(target)

# result = np.delete(mat, 4, 1)
result = np.reshape(matrix[mask], (-1, 4))

{'result after removing target': result}
```

```python
# Mean | Median | standard deviation

mean = np.mean(result[:, 0])
median = np.median(result[:, 0])
sd = np.std(result[:,0])

{
    'mean': mean,
    'median': median,
    'sd': sd
}
```

```python
# Set np.nan on random positions

iris_arr = result.copy()

rand_i = np.random.choice(150, size=20, replace=False)
rand_j = np.random.choice(4, size=20)

iris_arr[rand_i, rand_j] = np.nan

{ 'matrix after setting nans': iris_arr }
```

```python
# Find the positions of the inserted np.nan values in the 1st column

# Position in 2d matrix
# Obviously the i unit vector is 0, but I decided to play it safe
nan_mat = np.isnan(iris_arr)
nan_mat[:,1:] = False
nans_first = np.where(nan_mat == True)

# Position in firsts column
nans_first_col = np.where(np.isnan(iris_arr[:,0]))

{ 'positions of nans in first column': nans_first_col }
```

```python
# Filter array

filtered_arr = iris_arr[(iris_arr[:, 2] > 1.5) & (iris_arr[:, 0] < 5.0)]
{'filtered array': filtered_arr}
```

```python
# Set every np.nan value to 0

iris_arr[np.isnan(iris_arr)] = 0

{ 'matrix after setting zeros': iris_arr }
```

```python
# Get unique values and count them

unique = np.unique(iris_arr)

{'array': unique, 'amount of unique values': len(unique)}
```

```python
# Vertically split array

# two_arrays = np.split(mat_removed, 2, 0)
two_arrays = np.array_split(iris_arr, 2, 0)
[first_array, second_array] = two_arrays

{'two arrays': two_arrays}
```

```python
# Sort those two arrays

first_sorted = first_array[first_array[:,0].argsort()]
second_sorted = second_array[second_array[:,0].argsort()[::-1]]

{'first sorted array': first_sorted, 'second sorted array': second_sorted}
```

```python
# Join arrays into one

sorted_join = np.concatenate((first_sorted, second_sorted), axis=0)

{'sorted join': sorted_join}
```

```python
# Find the most frequent value

values, counts = np.unique(sorted_join, return_counts=True)
most_frequent = values[counts.argmax()]

{'most frequent value': most_frequent}
```

```python
# Define function

def transform_column(col):
    avg = np.average(col)

    cla = col < avg

    col[cla] *= 2
    col[~(cla)] /= 4
```

```python
# Apply the obtained function to the 3rd column

transform_column(sorted_join[:,2])
{'resulted array': sorted_join}
```

```python
# Make copy, reverse it vertically, transpose it, dot multiply them

reversed_sorted = np.flip(sorted_join, axis=0)
reversed_transposed = np.transpose(reversed_sorted)

multiply = np.dot(reversed_sorted, reversed_transposed)

{ 'result of multiplication': multiply, 'shape': multiply.shape }
```

```python
# Check if the matrix is singular

# to apply operations to a matrix, we need to cast the values to an integer typeTo to apply operations to a matrix,
matrix = multiply.astype(int)

determinant = np.linalg.det(matrix)

mat_is_sing = determinant == 0

{'matrix is singular': mat_is_sing }
```

```python
# Find the inverse matrix for our newly foundFind inverse matrix for

copied_mat = matrix.copy()
# Because the matrix is singular (sometimes), there is no inverse matrix.
# so let's make it non-singular (for cases it is singular)
# copied_mat[-1,:] *= 8

inv = np.linalg.inv(copied_mat)

{'inverse matrix': inv}
```
